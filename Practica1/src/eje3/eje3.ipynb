```python
import lecturas as l
import perceptron_adaline as pa
import classes_neuro as cn
```

## Problemas Logicos con Perceptron y adaline

- ### Perceptron

```python
texts = ["and.txt","or.txt","nand.txt"]
threshold = 0.2
for txt in texts:
    inputs,expected_outputs=l.read2(f"../texts/" + txt)
    network, input_layer, output_layer = pa.init_monolayer_nn(len(inputs[0]), len(expected_outputs[0]), "perceptron", 1,
                                                           threshold)
    mse_list,accuracies = pa.train_perceptron(network, input_layer, output_layer,inputs, expected_outputs, 1000)
    cn.plot_list(mse_list,"Epoch","MSE","MSE Development")
    cn.plot_list(accuracies,"Epoch","Accuracy","Accuracy Development")
```

- ### Adaline

```python
texts = ["and.txt","or.txt","nand.txt"]
for txt in texts:
    inputs,expected_outputs=l.read2(f"../texts/" + txt)
    threshold = 0.2
    network, input_layer, output_layer = pa.init_monolayer_nn(len(inputs[0]), len(expected_outputs[0]), "adaline", 1,
                                                           threshold)
    mse_list,accuracies = pa.train_adaline(network, input_layer, output_layer,inputs, expected_outputs,threshold, 1000)
    cn.plot_list(mse_list,"Epoch","MSE","MSE Development")
    cn.plot_list(accuracies,"Epoch","Accuracy","Accuracy Development")
```
